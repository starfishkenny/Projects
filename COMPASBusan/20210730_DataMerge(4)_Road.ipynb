```python
# import libraries

import os
from pathlib import Path
import sys
import time

import pandas as pd
import numpy
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt


import folium
import geopandas as gpd
import fiona
from shapely.geometry import Point, LineString
import pyproj

import json

%matplotlib inline


print(sys.version)
```

```python
# 도로 데이터 불러오기

addr_road = gpd.read_file('data/11.부산시남구_도로명주소(도로).geojson')
addr_road.head(5)
```

```python
# 고속도로 있는 지 확인

addr_road['ROA_CLS_SE'].value_counts() # 고속도로 없음
```

```python
# 도로 위계 구분에 따라 분리하여 지도에 나타내기

addr_road.to_crs(epsg='4326')
addr_road_json = addr_road.to_json()

loc = [35.1300,129.100]

map_road = folium.Map(location=loc, tiles = 'cartodbpositron', zoom_start=13)

style_function = lambda x: {'color':'#000000' if x['properties']['ROA_CLS_SE']=='2' else\
                            '#88969e' if x['properties']['ROA_CLS_SE']=='3' else\
                            '#9c7b7b',
                            'weight': 1}

folium.GeoJson(
    addr_road_json,
    style_function = style_function,
   ).add_to(map_road)

map_road
```
